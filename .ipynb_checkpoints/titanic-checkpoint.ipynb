```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
%matplotlib inline
```

```python
train = pd.read_csv('train.csv')

#fill NaNs
avg_age = np.mean(train['Age'])
train['Age'] = train['Age'].fillna(avg_age)
train['Cabin'] = train['Cabin'].fillna(False)

#normalize 'Age' and 'Fare' columns
fare_array = np.array(train['Fare'].values)
fare_array = fare_array.reshape(-1,1)
norm_fare = preprocessing.normalize(fare_array, axis=0)
train['Fare'] = norm_fare
age_array = np.array(train['Age'].values)
age_array = age_array.reshape(-1,1)
norm_age = preprocessing.normalize(age_array, axis=0)
train['Age'] = norm_age

#change 'cabin' and 'Sex' into numerical data types
train['cabin_bool'] = [(x != False) for x in train['Cabin'].values]
train['Sex'] = [(x == 'male') for x in train['Sex'].values]
print(train['Survived'].value_counts())
train.head(45)
```

```python
#define and split training variables
X = train[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'cabin_bool']]
y = train['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=20)

#instantiate model and fit data
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
prediction_test = model.predict(X_test)

#inspect results
print(metrics.accuracy_score(y_test, prediction_test))
```
